Question 4: You are setting up a web server on a Linux machine. You run your server script but
it says 'Permission denied'. Your senior tells you the file needs to be executable. What will you
do to fix this?

**ANSWER-** I will use the "chmod" command to grant execution permission to the script. Specifically, I would run:
"chmod +x script_name.sh"
The "chmod" command is used to change the access permissions of a file. Which allows the system to run the file as a program or script.
Linux blocks the file from running for security reasons, resulting in the "Permission denied" error.

Question 8: You are building an online store page for a shop in your city. The page has a
product listing, a contact form, a header with the shop name, and a footer with address. Your
senior says you should use semantic HTML tags. Which tags will you use for each section and
why?

**ANSWER-** I would use the following semantic tags to provide meaning and structure  to the page:

* header: For the shop name and navigation, as it defines the inductory section of the page.

*section: For the product listing to group related items together.

*form: For inside a section for the contact form to define interactive controls.

*footer: For the address and copyright info, as it represents the closing section of the page.

*address: Inside the footer to specifically mark the contact details for the shop.

-**Why**? Using semantic tags improves **SEO**.
It enhances accessibility for screen readers, and makes the code much easier for other developers to read and maintain.

Question 11: What is the difference between semantic HTML tags like section and article
versus a generic div tag? Why does it matter?

**ANSWER-** The primary difference is meaning (semantics). A div is a generic container with no meaning, used mainly for layout and styling. In contrast, section and article are semantic tags that tell the browser and search engines exactly.

It matters because:

* Accessibility: It helps screen readers navigate the page for users with disabilities.

* SEO: Search engines can better index the site by identifying the most important parts of the content.

* Maintainability: It makes the code cleaner and easier for developers to understand the structure without relying solely on class names..

Question 12: Why are forms important in web development? Name three input types used in
HTML forms and explain what each is used for.

**ANSWER-** Forms are the primary way a website collects data from a user. Without forms, the internet would be "read-only." They allow for:

* User Interaction: Creating accounts, logging in, and posting comments.

* E-commerce: Entering shipping addresses and payment details.

* Communication: Sending messages through contact sections.

* Data Processing: Searching for information on a site or filtering product results.
